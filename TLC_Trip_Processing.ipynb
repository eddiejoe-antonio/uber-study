```python
# Import aggregate report for number of unique drivers
df_all = pd.read_csv('./FHV_Base_Aggregate_Report_20240809.csv')

# Filter the DataFrame for rows where 'Base License Number' is 'UBER'
# and 'Year' is 2023 or 2024
df_uber = df_all[(df_all['Base License Number'] == 'UBER') & (df_all['Year'].isin([2023, 2024]))]
df_uber
```

```python
# Sort the filtered DataFrame by 'Year' (ascending) and 'Month' (ascending)
df_sorted = df_uber.sort_values(by=['Year', 'Month'], ascending=[True, True])

# Filter for the range April 2023 to March 2024
df_filtered = df_sorted[((df_sorted['Year'] == 2023) & (df_sorted['Month'] >= 4)) |
                        ((df_sorted['Year'] == 2024) & (df_sorted['Month'] <= 3))]

# Display the filtered DataFrame
df_filtered
```

```python
# Calculate mean monthly unique drivers
unique_monthly_drivers = df_filtered['Unique Dispatched Vehicles'].mean()
unique_monthly_drivers

# Initialize a base list
results = []
print(results)
```

```python
# List of monthly Parquet files
parquet_files = [
    './tlc_data/fhvhv_tripdata_2023-04.parquet',
    './tlc_data/fhvhv_tripdata_2023-05.parquet',
    './tlc_data/fhvhv_tripdata_2023-06.parquet',
    './tlc_data/fhvhv_tripdata_2023-07.parquet',
    './tlc_data/fhvhv_tripdata_2023-08.parquet',
    './tlc_data/fhvhv_tripdata_2023-09.parquet',
    './tlc_data/fhvhv_tripdata_2023-10.parquet',
    './tlc_data/fhvhv_tripdata_2023-11.parquet',
    './tlc_data/fhvhv_tripdata_2023-12.parquet',
    './tlc_data/fhvhv_tripdata_2024-01.parquet',
    './tlc_data/fhvhv_tripdata_2024-02.parquet',
    './tlc_data/fhvhv_tripdata_2024-03.parquet',
]

# Initialize accumulators for the sums (all, 'WAV', and 'Non-WAV')
totals = {
    "all": {"trip_miles": 0, "drivers": 0, "driver_pay_notips": 0, "tips": 0},
    "WAV": {"trip_miles": 0, "drivers": 0, "driver_pay_notips": 0, "tips": 0},
    "Non-WAV": {"trip_miles": 0, "drivers": 0, "driver_pay_notips": 0, "tips": 0},
}

# Loop through each file and sum the relevant columns
for file in parquet_files:
    # Load the required columns
    df = pd.read_parquet(file, columns=['trip_miles', 'driver_pay', 'tips', 'wav_request_flag'])
    
    # Process totals for all records
    totals["all"]["trip_miles"] += df['trip_miles'].sum()
    totals["all"]["driver_pay_notips"] += df['driver_pay'].sum()
    totals["all"]["tips"] += df['tips'].sum()
    
    # Process totals where 'wav_request_flag' == 'Y' (WAV)
    df_wav = df[df['wav_request_flag'] == 'Y']
    totals["WAV"]["trip_miles"] += df_wav['trip_miles'].sum()
    totals["WAV"]["driver_pay_notips"] += df_wav['driver_pay'].sum()
    totals["WAV"]["tips"] += df_wav['tips'].sum()
    
    # Process totals where 'wav_request_flag' == 'N' (Non-WAV)
    df_non_wav = df[df['wav_request_flag'] == 'N']
    totals["Non-WAV"]["trip_miles"] += df_non_wav['trip_miles'].sum()
    totals["Non-WAV"]["driver_pay_notips"] += df_non_wav['driver_pay'].sum()
    totals["Non-WAV"]["tips"] += df_non_wav['tips'].sum()

# Calculate total pay for each category
for key in totals:
    totals[key]["total_pay"] = totals[key]["driver_pay_notips"] + totals[key]["tips"]

# Append results for all records to the results list
results.append(["Total Trip Miles (All)", totals['all']['trip_miles']])
results.append(["Mean Monthly Unique Drivers (All)", unique_monthly_drivers])
results.append(["Total Driver Pay Before Tips (All)", totals['all']['driver_pay_notips']])
results.append(["Total Tips (All)", totals['all']['tips']])
results.append(["Total Pay (All)", totals['all']['total_pay']])

# Append results for 'wav_request_flag' == 'Y' (WAV)
results.append(["Total Trip Miles (WAV)", totals['WAV']['trip_miles']])
results.append(["Total Driver Pay Before Tips (WAV)", totals['WAV']['driver_pay_notips']])
results.append(["Total Tips (WAV)", totals['WAV']['tips']])
results.append(["Total Pay (WAV)", totals['WAV']['total_pay']])

# Append results for 'wav_request_flag' == 'N' (Non-WAV)
results.append(["Total Trip Miles (Non-WAV)", totals['Non-WAV']['trip_miles']])
results.append(["Total Driver Pay Before Tips (Non-WAV)", totals['Non-WAV']['driver_pay_notips']])
results.append(["Total Tips (Non-WAV)", totals['Non-WAV']['tips']])
results.append(["Total Pay (Non-WAV)", totals['Non-WAV']['total_pay']])
print(results)
```

```python
# Calculate weekly mileage for all drivers
mean_weekly_mileage = totals['all']['trip_miles']/unique_monthly_drivers/52
print("Mean weekly mileage for all drivers: ", mean_weekly_mileage)
results.append(["Mean Weekly Mileage", mean_weekly_mileage])

# Calculate weekly pay for all drivers
mean_weekly_salary = totals['all']['driver_pay_notips']/unique_monthly_drivers/52
print("Mean weekly salary for all drivers: $", mean_weekly_salary)
results.append(["Mean Weekly Salary", mean_weekly_salary])

# Calculate weekly pay for all drivers
mean_weekly_tips = totals['all']['tips']/unique_monthly_drivers/52
print("Mean weekly pay for all drivers: $", mean_weekly_tips)
results.append(["Mean Weekly Tips", mean_weekly_tips])

# Calculate weekly pay for all drivers
mean_weekly_pay = totals['all']['total_pay']/unique_monthly_drivers/52
print("Mean weekly pay for all drivers: $", mean_weekly_pay)
results.append(["Mean Weekly Total Earnings", mean_weekly_pay])

# Calculate annual pay for all drivers
mean_annual_pay = totals['all']['total_pay']/unique_monthly_drivers
print("Mean annual pay for all drivers: $", mean_annual_pay)
results.append(["Mean Annual Pay", mean_annual_pay])
```

```python
# Create a DataFrame from the results list
df_results = pd.DataFrame(results, columns=['Metric', 'Value'])
df_results['Value'] = pd.to_numeric(df_results['Value'], errors='coerce')

# Export the results 
df_results.to_csv('./tlc_results.csv')
```
